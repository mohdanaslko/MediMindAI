```python
columns = df.columns
column = columns.drop('Disease')
symptoms = []
for i in column:
    values = df[i].unique()
    for val in values:
        if val not in symptoms and pd.notnull(val):
            symptoms.append(val.strip().replace('_', ' '))

print(symptoms)
```

```python
dataset = pd.DataFrame(0,index = df.index, columns=symptoms)
dataset 
```

```python
for i in range(len(df)):
    for j in column:
        if pd.notnull(df[j][i]):
            dataset.at[i, df[j][i].strip().replace('_', ' ')] = 1
dataset['Disease'] = df['Disease']
dataset.head()
```
